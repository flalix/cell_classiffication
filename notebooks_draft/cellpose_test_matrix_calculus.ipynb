### create batched Q,K like attention uses

```python
B, H, N, D = 2, 4, 128, 64   # small-ish test sizes
q = torch.randn(B, H, N, D, device='cuda', dtype=torch.float32)
k = torch.randn(B, H, N, D, device='cuda', dtype=torch.float32)
```

### scaled matmul that attention uses

```python
scale = (D**-0.5)
try:
    out = (q * scale) @ k.transpose(-2, -1)
    print("FP32 batched matmul OK:", out.shape)
except Exception as e:
    print("ERROR:", e)
```

### Error?
  - If this fails, PyTorch is still choosing an unsupported cuBLAS path for batched GEMM.
  - If this succeeds, the problem is likely mixed precision / autocast inside Cellpose/model code.
